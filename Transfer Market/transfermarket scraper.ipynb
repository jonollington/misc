```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
```

```python
page = "https://www.transfermarkt.co.uk/arsenal-fc/kader/verein/11/saison_id/2019/plus/1"
pageTree = requests.get(page, headers=headers)
pageSoup = BeautifulSoup(pageTree.content, 'html.parser')
```

```python
ages = pageSoup.findAll("a", {"class" : "zentriert"})
[(age.getText(), age.attrs['href']) for age in ages]
```

```python
players = pageSoup.findAll("a", {"class" : "spielprofil_tooltip"})
[(player.getText(), player.attrs['href']) for player in players]
```

```python
#Process each of the players lists into text-only lists
PlayersText = []
for player in players:
    PlayersText.append(player.text)
```

```python
PlayersText
```

```python
player=[player for player in PlayersText][::2]
```

```python
player
```

```python
signed_from = pageSoup.findAll("a", {"class" : "vereinprofil_tooltip"})
signed_from_data = [(sign.img.attrs['alt'], sign.img.attrs['title']) for sign in signed_from ]
```

```python
signed_from_data
```

```python
signed_from_text = []
for team in signed_from_data:
    signed_from_text.append(team[:][0])
```
