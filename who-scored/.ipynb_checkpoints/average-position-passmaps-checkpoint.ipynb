```python
class Player:
    def __init__(self, player, df):
        self.id = player["player"]["id"]
        self.name = player["player"]["name"]
        self.average_position(df)

    def average_position(self, df):

        player_pass_df = df.query("(Event == 'Pass') & (passtype not in ['CornerTaken', 'Throwin']) & (player_id == @self.id) & (pass_outcome_name not in ['Unknown','Out','Pass Offside','Injury Clearance', 'Incomplete'])")
        self.x = np.mean(player_pass_df['X'].tolist(), axis=0)
        self.y = np.mean(player_pass_df['Y'].tolist(), axis=0)

        self.n_passes_completed = len(player_pass_df)
```
