```python
df = pd.read_csv('./mid_pressures.csv')
pd.set_option('display.max_columns', None)
```

```python
df['90s'] = pd.to_numeric(df['90s'])

df['90s'] = df['90s'].apply(pd.to_numeric, errors='coerce')
```

```python
df.
```

```python
mask = (df.Comp == 'eng Premier League') & (df.Pos == 'MF') & (df['90s'] > 5.5)
df=df.loc[mask]
df.head()
```

```python
pressure = df.iloc[:, 18:24]
```

```python
pressure
```

```python
master_df = pd.concat([df[['Player', 'Squad','90s']], pressure], axis = 1)
master_df
```

```python
master_df['90s'] = pd.to_numeric(master_df['90s'])
master_df['Succ'] = pd.to_numeric(master_df['Succ'])
master_df['Att 3rd'] = pd.to_numeric(master_df['Att 3rd.1'])
master_df['Mid 3rd'] = pd.to_numeric(master_df['Mid 3rd.1'])
master_df['Def 3rd'] = pd.to_numeric(master_df['Def 3rd.1'])
```

```python
master_df['Succ p90'] = master_df['Succ'] / master_df['90s']
master_df['Def 3rd p90'] = master_df['Def 3rd'] / master_df['90s']
master_df['Mid 3rd p90'] = master_df['Mid 3rd'] / master_df['90s']
master_df['Att 3rd p90'] = master_df['Att 3rd'] / master_df['90s']
```

```python
master_df
```

```python
master_df=master_df.dropna()
```

```python
len(master_df)
```

```python
master_df.dtypes
```

```python
# Drop columns
master_df1 = master_df.drop(['Player','Squad','90s','Press','Succ','%','Def 3rd','Mid 3rd','Att 3rd','Def 3rd.1','Mid 3rd.1','Att 3rd.1'], axis=1)
```

```python
master_df1
```

```python
len(master_df1)
```

```python
# Normalize data
DFNom = master_df1.transform(lambda x: (x - x.min())/ (x.max() - x.min()))
```
