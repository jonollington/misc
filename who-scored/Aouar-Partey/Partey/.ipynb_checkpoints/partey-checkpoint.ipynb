```python
#calculate pass distance
aouar['dist'] = np.sqrt( (aouar.endX-aouar.x)**2 + (aouar.endY-aouar.y)**2)
```

```python
aouar['end_dist_from_goal'] = np.sqrt( (aouar.endX-100)**2 + (aouar.endY-50)**2)
```

```python
aouar['start_dist_from_goal'] = np.sqrt( (aouar.x-100)**2 + (aouar.y-50)**2)
```

```python
aouar['prog_pass'] = aouar['start_dist_from_goal'] - aouar['end_dist_from_goal']
```
