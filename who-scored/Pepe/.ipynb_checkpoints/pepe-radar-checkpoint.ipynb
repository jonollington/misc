```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as patches
from math import pi

%matplotlib inline
```

```python
df = pd.read_csv('./2019-2020 Big 5 Euro PCA Stats.csv')
pd.set_option('display.max_columns', None)
```

```python
# Remove non-numerical data
all_columns = list(df.columns)
unwanted_columns = {'Rk','Player', 'Nation','Pos', 'Squad', 'Comp', 'Age','Born','90s'}
index_columns = [ele for ele in all_columns if ele not in unwanted_columns]
```

```python
df1 = df.apply(pd.to_numeric, errors='coerce')
```

```python
dfP90 = df1.loc[:,index_columns].div(df1['90s'], axis=0)
```

```python
dfP90.fillna(0, inplace=True)
```

```python
# P90 Dataframe
master_df = pd.concat([df[['Player', 'Nation', 'Pos', 'Squad', 'Comp', 'Age', 'Born', '90s']], dfP90], axis = 1)
```

```python
master_df['90s'] = master_df['90s'].apply(pd.to_numeric, errors='coerce')
```

```python
master_df['Age'] = master_df['Age'].apply(pd.to_numeric, errors='coerce')
```

```python
# Filter for players who have played at least 1000 minutes & remove Goalkeepers
mask0 = (master_df['90s'] > 11.11) & ((master_df.Pos == 'FW') | (master_df.Pos == 'FW,MF')
                                    | (master_df.Pos == 'MF,FW'))
master_df = master_df.loc[mask0]
```

```python
mask=(master_df['Comp']=='eng Premier League') | (master_df['Comp']=='it Serie A') & (master_df['Player']=='Lorenzo Insigne')
master_df = master_df[mask]
master_df
```

```python
mask2=master_df['Player']=='Lorenzo Insigne'
master_df[mask2]
```

```python
# Drop columns
df1 = master_df.drop(['Player', 'Nation','Pos', 'Squad', 'Comp', 'Age', 'Born', '90s'], axis=1)
```

```python
DFNom=df1.rank(pct=True)
```

```python
# #Normalize data
# DFNom = df1.transform(lambda x: (x - x.min())/ (x.max() - x.min()))
```

```python
master_df = pd.concat([master_df[['Player','Pos','Squad','Comp','Age']], DFNom], axis = 1)
master_df
```

```python
master_df = master_df[['Player','xG','passPrgDist','Ast','xA','KP',
                  'PPA','CrsPA','TB','SCAPassLive','Drib Succ', 'PrgDist']]
```

```python
master_df.rename(columns={
    'CrsPA':'Crosses',
    'Ast':'Assists',
    'PPA':'Passes into Penalty Area',
    'TB':'Through Balls',
    'Drib Succ':'Successful Dribbles',
    'PrgDist':'Prog. Dist. Carries',
    'passPrgDist':'Prog. Dist. Passes',
    'SCAPassLive':'Shot Creating Actions',
    'KP':'Key Passes',
    
}, inplace=True)
```

```python
def createRadar(player, player2):

    # Filter df 
    playerdf = master_df[master_df['Player'] == player]
    
    if playerdf.empty == True:
        print('DF Empty')
        return
    
    playerdf2 = master_df[master_df['Player']==player2]
    
    if playerdf2.empty == True:
        print('DF Empty')
        return
    
    attributes = list(playerdf)[1:]
    values = playerdf.iloc[0,1:].tolist()
    values += values [:1]
    
    attributes2 = list(playerdf2)[1:]
    values2 = playerdf2.iloc[0,1:].tolist()
    values2 += values2 [:1]
  
    angles = [n / 11 * 2 * pi for n in range(11)]
    angles += angles [:1]
    
    angles2 = [n / 11 * 2 * pi for n in range(11)]
    angles2 += angles2 [:1]
    
    # Add Plot
    fig, ax = plt.subplots(subplot_kw=dict(projection='polar'),figsize=(8,8))
    plt.rcParams["font.family"] = "BBC Reith Sans"
    
    plt.xticks(angles[:-1],attributes, )
    
    ax.plot(angles,values, 'firebrick', linewidth=.3, alpha=0.2,zorder=10)
    ax.fill(angles,values, 'firebrick', alpha=0.6)
    
    ax.plot(angles2,values2, 'skyblue', linewidth=.3, alpha=0.2,zorder=10)
    ax.fill(angles2, values2, 'skyblue', alpha=0.5)

    
    # Add rotated Labels
    sections = len(attributes)
    section_width = np.deg2rad(360/sections)
    values = master_df.iloc[0,3:].tolist()

    for i in range (0,sections):
        x = (section_width*i)
        y = 1.05

        rot = np.rad2deg(x)-90

        ax.text(x, y, attributes[i], alpha=1, ha="center", va="center", rotation=rot, fontsize=12, color='white')
    
    #Set position of y-labels (0-100) to be in the middle of the first two axes.
    ax.set_rlabel_position(84)
    ax.grid(color='lightgrey',zorder=0,alpha=0.09)
    
    # Change the color of the outermost gridline (the spine).
    ax.spines['polar'].set_color(None)
    
    # Make the y-axis (0-1) labels smaller.
    ax.tick_params(axis='y', labelsize=0, colors='white')
    yt = [20,40,60,80,100]
    ax.set_yticklabels(yt)
    
    labels = [item.get_text() for item in ax.get_xticklabels()]

    empty_string_labels = ['']*len(labels)
    ax.set_xticklabels(empty_string_labels)
    
    # Ensure radar goes from 0 to 100.
    ax.set_ylim(0, 1)

    
    circle = plt.Circle((0, 0), 0.9, transform=ax.transData._b, fill=False, edgecolor='#AAAAAA', linewidth=46, zorder=0,alpha=0.4)
    circle1 = plt.Circle((0, 0), 0.5, transform=ax.transData._b, fill=False, edgecolor='#AAAAAA', linewidth=46, zorder=0, alpha=0.4)
    circle2 = plt.Circle((0, 0), 0.1, transform=ax.transData._b, fill=False, edgecolor='#AAAAAA', linewidth=46, zorder=0, alpha=0.4)


    plt.gca().add_artist(circle)
    plt.gca().add_artist(circle1)
    plt.gca().add_artist(circle2)
    
    plt.savefig("Willian_radar", dip=2000, bbox_inches='tight',transparent=True)
    plt.tight_layout()
```

```python
createRadar('Willian','Lorenzo Insigne')
```
