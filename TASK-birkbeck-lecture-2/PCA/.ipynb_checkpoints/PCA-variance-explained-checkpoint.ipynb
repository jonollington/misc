```python
df = pd.read_csv('./data/EPL 19-20 outfield openplay stats per 90_min500.csv')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Drop columns
df = df.drop(['position', 'player', 'team', 'foot'], axis=1)
```

```python
from sklearn import preprocessing
data_scaled = pd.DataFrame(preprocessing.scale(df),columns = df.columns) 
```

```python
# PCA
pca = PCA(n_components=2)
pca.fit_transform(data_scaled)
```

```python
train_features = pd.DataFrame(preprocessing.scale(df),columns = df.columns)
```

```python
model = PCA(n_components=2).fit(train_features)
X_pc = model.transform(train_features)
```

```python
n_pcs= model.components_.shape[0]
```

```python
# get the index of the most important feature on EACH component i.e. largest absolute value
# using LIST COMPREHENSION HERE
most_important = [np.abs(model.components_[i]).argmax() for i in range(n_pcs)]
```

```python
most_important
```

```python
initial_feature_names = [x for x in df.columns]
```

```python
# get the names
most_important_names = [initial_feature_names[most_important[i]] for i in range(n_pcs)]
```

```python
# using LIST COMPREHENSION HERE AGAIN
dic = {'PC{}'.format(i+1): most_important_names[i] for i in range(n_pcs)}
```

```python
# build the dataframe
df = pd.DataFrame(sorted(dic.items()))
```
