```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df = pd.read_csv('../data/player_game_statistic_10_2019.csv')
pd.set_option('display.max_columns', None)
```

```python
df
```

```python
# Filter for players who have played at least 1000 minutes & remove Goalkeepers
mask = (df.mins_played > 500) & (df.position != 'Goalkeeper')
df = df.loc[mask]
```

```python
# Check relevant players have been removed
df.position.value_counts()
```

```python
# Create a new column/metric for number of 90 minutes played
df['nineties'] = df['mins_played'] / 90
df.head()
```

```python
# Remove non-numerical data
all_columns = list(df.columns)
unwanted_columns = {'player_id','position','player','age','team_id','team','foot','mins_played','nineties'}
index_columns = [ele for ele in all_columns if ele not in unwanted_columns]
```

```python
# create new dataframe based on PP90mins
dfP90 = df.loc[:,index_columns].div(df['nineties'], axis=0)
master_df = pd.concat([df[['player_id','position','player','age','team_id','team','foot','mins_played','nineties']], dfP90], axis = 1)
master_df
```

```python
# Save dataframe
master_df.to_csv('EFL 19-20 outfied P90 500min.csv', index=False)
```
