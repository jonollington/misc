```python
players = df['player'].tolist()
```

```python
# Drop columns
df1 = df.drop(['position', 'player', 'team', 'foot'], axis=1)
```

```python
df1
```

```python
x = df1.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()  # could also test using the StandardScaler()
x_scaled = min_max_scaler.fit_transform(x)
X_norm = pd.DataFrame(x_scaled)
```

```python
pca = sklearnPCA(n_components=2) #2-dimensional PCA
transformed = pd.DataFrame(pca.fit_transform(X_norm))
```

```python
coefs = pd.DataFrame(pca.components_,columns=df1.columns,index = ['PC-1','PC-2'])
```

```python
coefs.to_csv('coefs.csv')
```

```python
coefs
```

```python
cluster_range = range( 1, 11 )
cluster_errors = []

for num_clusters in cluster_range:
  clusters = KMeans( num_clusters )
  clusters.fit( transformed )
  cluster_errors.append( clusters.inertia_ )
```

```python
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )
```

```python
sns.set(style="white")
plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
plt.tick_params(labelsize=15)

plt.xlabel("Clusters", fontsize=20)
plt.ylabel("Sum of squared errors", fontsize=20)
```

```python
# Number of clusters
kmeans = KMeans(n_clusters=5)
# Fitting the input data
kmeans = kmeans.fit(transformed)
# Getting the cluster labels
labels = kmeans.predict(transformed)
# Centroid values
C = kmeans.cluster_centers_
clusters = kmeans.labels_.tolist()
```

```python
new_df = transformed.rename(columns={0:'PC-1', 1: 'PC-2'})
new_df['Player'] = players
new_df['Clusters'] = clusters
```

```python
new_df.to_csv('all_data.csv')
```
